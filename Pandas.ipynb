```python
#Question 1
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'], 
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']}) 
df.loc[1,'FlightNumber'] = 10055                    # Assigning Values
df.loc[3,'FlightNumber'] = 10075                    # Assigning Values
df['FlightNumber'] = df['FlightNumber'].astype(int) # converting data into integers
df
```

```python
#Question 2
split_data = df['From_To'].str.split("_") # Splitting data
data = split_data.to_list()               # Return a list of the values
names = ['From', 'To']                    # Column Names
new_df = pd.DataFrame(data, columns=names)
new_df
```

```python
#Question 3
new_df['From'] = new_df['From'].str.capitalize()  #Capital Letters
new_df['To'] = new_df['To'].str.capitalize()      # Capital Letters
new_df
```

```python
# Question 4
df = df.drop('From_To',1)
new_df = pd.concat([new_df,df],axis = 1) #concatination of new_df and df
new_df                                    
```

```python
# Question 5
x = 1
d2 = (pd.DataFrame(new_df['RecentDelays'].values.tolist(),index=new_df.index).rename(columns = lambda x:'delay_{}'.format(x+1)))
new_df = new_df.drop('RecentDelays',1)  #Dropping RecentDelays Column
df2 = new_df.pop('Airline')             #popping Airline Column
df1 = new_df.join(d2)                   # join d2
df1['Airline'] = df2                    #Adding the Airline column to the end of DataFrame
df1                         
```
